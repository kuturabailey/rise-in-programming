```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data file
file_path = '/Users/kuturabailey/Downloads/ActionabilityData.pdf'
data = pd.read_csv(file_path, sep='\\t', error_bad_lines=False, engine='python')

# Filter based on keywords
keywords = ['EGFR', 'FGFR1', 'lung/carcinoma/non small cell carcinoma']
filtered_data = data[data['MUTATION_REMARK'].str.contains('|'.join(keywords), na=False)]

# Define the columns to include in the output
columns_to_keep = [
    'Gene', 'MUTATION_REMARK', 'Disease', 'ACTIONABILITY_RANK', 'DRUG_COMBINATION', 
    'Source of Info', 'Primary Outcome Measure'
]
filtered_data = filtered_data[columns_to_keep]

# Save filtered data to a new file
output_file = 'filtered_actionability_data.csv'
filtered_data.to_csv(output_file, index=False)

# Simplified parsed file
parsed_data = filtered_data[['Gene', 'Disease', 'MUTATION_REMARK', 'ACTIONABILITY_RANK']]
parsed_output_file = 'parsed_actionability_data.csv'
parsed_data.to_csv(parsed_output_file, index=False)

# Plot 1: Mutation and ACTIONABILITY_RANK
mutation_rank_plot = filtered_data.groupby('MUTATION_REMARK')['ACTIONABILITY_RANK'].value_counts()
mutation_rank_plot.unstack().plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Mutation and Actionability Rank')
plt.xlabel('Mutation')
plt.ylabel('Actionability Rank Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('mutation_actionability_rank_plot.png')
plt.close()

# Plot 2: Drug Combination and Primary Outcome Measure
drug_outcome_plot = filtered_data.groupby('DRUG_COMBINATION')['Primary Outcome Measure'].value_counts()
drug_outcome_plot.unstack().plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Drug Combination and Primary Outcome Measure')
plt.xlabel('Drug Combination')
plt.ylabel('Primary Outcome Measure Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('drug_combination_primary_outcome_plot.png')
plt.close()

print(f"Filtered data saved to {output_file}")
print(f"Parsed data saved to {parsed_output_file}")
print("Plots saved as mutation_actionability_rank_plot.png 
```
