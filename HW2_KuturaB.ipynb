```python
#Part V
#Definition of function to check gene sequence
def check_gene_sequence(sequence):
    # Define the valid nucleotide characters
    valid_nucleotides = ["A", "C", "G", "T"]
    
    # Initialize a list to store invalid characters
    invalid_chars = []
    
    # Iterate over the sequence using indexing
    for i in range(len(sequence)):
        if sequence[i] not in valid_nucleotides:
            invalid_chars.append(sequence[i])
    
    # Check if there are any invalid characters
    if invalid_chars:
        print("Invalid characters found: {invalid_chars}")
    else:
        # Calculate GC content if the sequence is valid
        gc_count = 0
        for nucleotide in sequence:
            if nucleotide in ['G', 'C']:
                gc_count += 1
        
        # Calculate GC content percentage
        gc_content_percentage = (gc_count / len(sequence)) * 100
        print(f"This sequence has {gc_content_percentage:.2f}% GC content.")

# Example sequence
gene_sequence = "AGCCCTCCAGGACAGGCTGCATCAGAAGAGGCCATCAAGC"
check_gene_sequence(gene_sequence)
```
