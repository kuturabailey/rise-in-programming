```python
import re

# Define regex patterns for ATP-binding pocket motifs
# Motif 1: GXGXXG(N/C) followed by any three amino acids, then (D/E)
motif_pattern_1 = r'G[A-Z]G[A-Z]{2}G[N|C][A-Z]{3}[D|E]'

# Motif 2: GXXXAGG
motif_pattern_2 = r'G[A-Z]{3}AGG'

# Function to find motifs in protein sequences using re.findall
def find_motifs(sequence, pattern):
    return re.findall(pattern, sequence)

# Read the fasta files and search for the motifs
def read_fasta_and_search(file_path):
    with open(file_path, 'r') as file:
        sequences = {}
        sequence_id = None
        sequence_data = []

        # Parse fasta format to retrieve sequence data
        for line in file:
            line = line.strip()
            if line.startswith(">"):  # Header line
                if sequence_id:
                    sequences[sequence_id] = ''.join(sequence_data)
                sequence_id = line
                sequence_data = []
            else:
                sequence_data.append(line)
        if sequence_id:
            sequences[sequence_id] = ''.join(sequence_data)  # Add last sequence

    # Search for motifs in each sequence
    results = {}
    for seq_id, seq in sequences.items():
        matches_1 = find_motifs(seq, motif_pattern_1)
        matches_2 = find_motifs(seq, motif_pattern_2)
        results[seq_id] = {
            "Motif GXGXXG(N/C)...(D/E)": matches_1,
            "Motif GXXXAGG": matches_2
        }
    return results

# Define file paths and search for motifs in each file
file_paths = [
    '/Users/kuturabailey/Downloads/ABL1_tyrkinase.fa',
    '/Users/kuturabailey/Downloads/TyrKin.fa'
]
for file_path in file_paths:
    motifs = read_fasta_and_search(file_path)
    print(f"Results for {file_path}:")
    for seq_id, match in motifs.items():
        print(seq_id)
        print("Matches for Motif GXGXXG(N/C)...(D/E):", match["Motif GXGXXG(N/C)...(D/E)"])
        print("Matches for Motif GXXXAGG:", match["Motif GXXXAGG"])
        print()
```
